```python
from wordcloud import WordCloud,STOPWORDS
```

```python
stopwords=set(STOPWORDS)
```

```python
join_tweets=' '.join(tweets['tweet'])
join_tweets=join_tweets.replace('https','.')
```

```python
from textblob import TextBlob
```

```python
polarity= []

for i in tweets['tweet']:
    polarity.append(TextBlob(i).sentiment.polarity)
```

```python
tweets['Polarity']=polarity
tweets.head(2)
```

```python
a = np.array(tweets['Polarity'])
```

```python
y=sum(np.array(a)>0)
z=sum(np.array(a)<0)
x=sum(np.array(a)==0)
p=z+y+x
```

```python
w=pd.DataFrame([x,y,z],index=['Positive','Negative','Neutral'])
w
```

```python
import plotly.express as px
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
labels=['Positive', 'Negative','Neutral']
```

```python
px.pie(w,values=0,labels=labels,names=['Positive', 'Negative','Neutral'],hole=0.5,title='Sentiment Spread for @aka_chineme', width=550, height=450.5)
```
